### Using ARIMA

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
# Ensure that the date index is in daily frequency
df_covid_global = df_covid_global.asfreq('D')

# Split the dataset
train_size = int(len(df_covid_global) * 0.8)
train, test = df_covid_global[['confirmed_growth_rate', 'death_growth_rate']][:train_size], df_covid_global[['confirmed_growth_rate', 'death_growth_rate']][train_size:]


# Function to fit ARIMA model and calculate RMSE
def fit_arima_and_evaluate(train, test, column, p, d, q):
    # Fit ARIMA model (you may need to adjust the (p, d, q) parameters)
    model = ARIMA(train[column], order=(p, d, q))  # Example order (1,1,1)
    model_fit = model.fit()

    # Make predictions
    predictions = model_fit.forecast(steps=len(test))
    
    # Calculate RMSE
    rmse = np.sqrt(mean_squared_error(test[column], predictions))
    
    return rmse, predictions
```

```python
# Fit ARIMA for confirmed_growth_rate and calculate RMSE
confirmed_rmse, confirmed_predictions = fit_arima_and_evaluate(train, test, 'confirmed_growth_rate', p = 2, d = 2, q = 2)

# Fit ARIMA for death_growth_rate and calculate RMSE
death_rmse, death_predictions = fit_arima_and_evaluate(train, test, 'death_growth_rate', p = 2, d = 1, q = 2)

# Output the RMSE results
print(f"Confirmed Growth Rate RMSE: {confirmed_rmse:.4f}")
print(f"Death Growth Rate RMSE: {death_rmse:.4f}")
```
