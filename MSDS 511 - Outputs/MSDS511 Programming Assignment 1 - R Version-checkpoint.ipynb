### Autocorrelation

```r
# Create a function to plot autocorrelation
plot_acf <- function(data, title, color) {
  acf_values <- acf(data, plot = FALSE, lag.max = 10)$acf[2:11]  # Calculate autocorrelation values
  lags <- 1:10
  
  # Create a data frame for plotting
  acf_df <- data.frame(lag = lags, value = acf_values)
  
  # Plot using ggplot2
  p <- ggplot(acf_df, aes(x = lag, y = value)) +
    geom_point(color = color) +
    geom_line(color = color) +
    geom_text(aes(label = round(value, 2)), vjust = -0.5, size = 3, color = color) +
    labs(title = title, x = 'Lags', y = 'Autocorrelation') +
    theme_minimal() +
    theme(
      plot.title = element_text(face = "bold"),
      panel.grid.major = element_line(size = 0.5, linetype = 'dashed'),
      panel.grid.minor = element_blank(),
      axis.title = element_text(face = "bold"),
      axis.ticks.length = unit(0.25, "cm")
    )
  
  return(p)
}

# Create individual plots for confirmed and death growth rates
confirmed_growth_rate_clean <- na.omit(df_covid_global$confirmed_growth_rate)
death_growth_rate_clean <- na.omit(df_covid_global$death_growth_rate)

# Create the plots
confirmed_plot <- plot_acf(confirmed_growth_rate_clean, 'Autocorrelation - Confirmed Growth Rate', 'blue')
death_plot <- plot_acf(death_growth_rate_clean, 'Autocorrelation - Death Growth Rate', 'red')
```

### Using ARIMA

```r
# Load necessary libraries
library(forecast)

# Ensure that the date index is in daily frequency (if your data is not already a time series object)
df_covid_global <- df_covid_global %>%
  mutate(date = as.Date(date)) %>%  # Ensure date is in Date format if not already
  arrange(date) %>%  # Sort by date
  as.ts()  # Convert to time series object

# Split the dataset into train and test sets
train_size <- floor(0.8 * nrow(df_covid_global))
train <- df_covid_global[1:train_size, c("confirmed_growth_rate", "death_growth_rate")]
test <- df_covid_global[(train_size + 1):nrow(df_covid_global), c("confirmed_growth_rate", "death_growth_rate")]

# Function to fit ARIMA model and calculate RMSE
fit_arima_and_evaluate <- function(train, test, column, p, d, q) {
  # Fit ARIMA model
  model <- Arima(train[[column]], order = c(p, d, q))
  
  # Make predictions
  predictions <- forecast(model, h = nrow(test))$mean
  
  # Calculate RMSE
  rmse <- sqrt(mean((test[[column]] - predictions)^2))
  
  return(list(rmse = rmse, predictions = predictions))
}
```

```r
# Fit ARIMA for confirmed_growth_rate and calculate RMSE
confirmed_results <- fit_arima_and_evaluate(train, test, "confirmed_growth_rate", p = 2, d = 2, q = 2)
confirmed_rmse <- confirmed_results$rmse
confirmed_predictions <- confirmed_results$predictions

# Fit ARIMA for death_growth_rate and calculate RMSE
death_results <- fit_arima_and_evaluate(train, test, "death_growth_rate", p = 2, d = 1, q = 2)
death_rmse <- death_results$rmse
death_predictions <- death_results$predictions

# Output the RMSE results
cat(sprintf("Confirmed Growth Rate RMSE: %.4f\n", confirmed_rmse))
cat(sprintf("Death Growth Rate RMSE: %.4f\n", death_rmse))
```

```r
# Load necessary libraries
library(forecast)

# Check column names to ensure they are correct
print(colnames(df_covid_global))

# Split the dataset into training and testing sets
train_size <- floor(0.8 * nrow(df_covid_global))

# Create train and test data frames
train <- as.data.frame(df_covid_global[1:train_size, c('confirmed_growth_rate', 'death_growth_rate')])
test <- as.data.frame(df_covid_global[(train_size + 1):nrow(df_covid_global), c('confirmed_growth_rate', 'death_growth_rate')])

# Check the structure of the data frames to ensure they are correctly created
str(train)
str(test)


# 1. Fit ARIMA model for confirmed_growth_rate
# Fit the ARIMA model
confirmed_model <- Arima(train$confirmed_growth_rate, order = c(2, 2, 2))

# Make predictions
confirmed_predictions <- forecast(confirmed_model, h = nrow(test))$mean

# Calculate RMSE
confirmed_rmse <- sqrt(mean((test$confirmed_growth_rate - confirmed_predictions) ^ 2))

# 2. Fit ARIMA model for death_growth_rate
# Fit the ARIMA model
death_model <- Arima(train$death_growth_rate, order = c(2, 1, 2))

# Make predictions
death_predictions <- forecast(death_model, h = nrow(test))$mean

# Calculate RMSE
death_rmse <- sqrt(mean((test$death_growth_rate - death_predictions) ^ 2))

# Output the RMSE results
cat(sprintf("Confirmed Growth Rate RMSE: %.4f\n", confirmed_rmse))
cat(sprintf("Death Growth Rate RMSE: %.4f\n", death_rmse))
```
